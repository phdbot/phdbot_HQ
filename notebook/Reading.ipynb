```python
words = text.split()  
```

```python
word_counts = Counter(words)
df = pd.DataFrame.from_dict(word_counts, orient='index')
```

```python
df.columns=['count']
```

```python
sorted_df = df.sort_values('count',ascending=False)
```

```python
sorted_df[len(sorted_df) - 100:]
```

```python
sorted_df[len(sorted_df) - 50:].plot(kind='bar')
```
